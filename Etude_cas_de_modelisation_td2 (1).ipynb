```python
def Intersection(x1, y1, theta1, x2, y2, theta2, l) :
  toto=np.matrix([[np.cos(theta1), -np.cos(theta2)], [np.sin(theta1), -np.sin(theta2)]])
  liwa=[x2-x1,y2-y1]
  if theta1 == theta2:
    return(False)
  p=np.linalg.solve(toto,liwa)
  if (0<=p[1]<=l) and (0<=p[0]<=l) :
    return(True)
  else : 
    return False
```

```python
# une fonction pour simuler loi de theta_j conditionnement  theta_j_1
def toto(n):
  alpha=0.5
  beta=7
  th0=3*np.pi/4
  th1=np.random.vonmises(th0,alpha+beta,n)
  zj=[]
  uj=[]
  kj=[]
  l=[]
  for a in np.arange(0,n):
    zj.append(alpha*np.cos(th0)+beta*np.cos(th1[a]) + j*(alpha*np.sin(th0)+beta*np.sin(th1[a])))
    uj.append(np.angle(zj[a]))
    kj.append(np.sqrt(alpha**2 + beta**2 +np.cos((th1[a])-th0)))
    l.append(np.random.vonmises(uj[a],kj[a],1))
  return l
```
